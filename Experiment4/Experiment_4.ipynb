```python
def plot_dft(x, fs,samples=None,l_label = None,clr = 'b') :
  X = np.abs(np.fft.fft(x[0:samples]))
  n = np.linspace(0, fs, len(X))
  plt.stem(n,X,clr,label = l_label)
  plt.legend()
  plt.title('Magnitude Spectrum Plot', fontsize = 20)
  plt.xlabel('Frequency', fontsize = 16)
  plt.ylabel('Amplitude', fontsize = 16)
  plt.legend(loc = 'upper right')
```

```python
plot_dft(y, fs = 120,samples= 120,l_label = "DFT with 120 samples")
plt.savefig('q1_p1.png')
```

```python
plot_dft(y, fs = 120,samples= 130,l_label = "DFT with 130 samples")
plt.savefig('q1_p2.png')
```

```python
plot_dft(y, fs = 120,samples= 120,clr = 'r',l_label = "DFT with 120 samples(Red)")
plot_dft(y, fs = 120,samples= 130,clr = 'g',l_label = "DFT with 130 samples(Green)")
plt.savefig('q1_p3.png')
```

There is no spectral leakage in DFT plotted for first 120 samples. But there is some spectral leakage for the DFT plotted for first 130 samples.

```python
def plot_dft(x, fs,samples=None,l_label = None,clr = 'b') :
  X = np.abs(np.fft.fft(x[0:samples]))
  n = np.linspace(0, fs, len(X))
  plt.stem(n,X / samples,clr,label = l_label)
  plt.legend()
  plt.title('Magnitude Spectrum Plot', fontsize = 20)
  plt.xlabel('Frequency in Hz', fontsize = 16)
  plt.ylabel('Normalized Amplitude', fontsize = 16)
  plt.legend(loc = 'upper right')
```

```python
plot_dft(y, fs = 120,samples= 120,clr = 'r',l_label = "DFT with 120 samples(Red)")
plot_dft(y, fs = 120,samples= 240,clr = 'g',l_label = "DFT with 240 samples(Green)")
plt.savefig('q1_p4.png')
```

For N = 240, or first 240 samples have the same DFT as the first 120 samples. There is no spectral leakage.

```python
# For alpha = 1
A = 120
B = 126
Fs = 200
t = np.arange(0,10,1/Fs)
x = 0.1 * np.sin(A * pi * t) + np.cos(B * pi * t)
```

```python
def plot_dft(x, fs,samples=None,l_label = None,clr = 'b') :
  X = np.abs(np.fft.fft(x[0:samples]))
  n = np.arange(len(X))/(len(X)/fs)
  plt.plot(n,X,clr,label = l_label)
  plt.legend()
  plt.title('Magnitude Spectrum Plot', fontsize = 20)
  plt.xlabel(r'$\Omega$', fontsize = 16)
  plt.ylabel(r'$|X(\Omega)|$', fontsize = 16)
  plt.legend(loc = 'upper right')
  plt.xlim(0, fs/2)
```

```python
# For alpha = 1
A = 120
B = 126
Fs = 200
t = np.arange(0,10,1/Fs)
x = 0.1 * np.sin(A * pi * t) + np.cos(B * pi * t)
```

```python
def plot_dft(x, fs,samples=None,l_label = None,clr = 'b') :
  X = np.abs(np.fft.fft(x[0:samples] * np.hamming(samples) ))
  n = np.arange(len(X))/(len(X)/fs)
  plt.plot(n,X,clr,label = l_label)
  plt.legend()
  plt.title('Magnitude Spectrum Plot applying hamming window', fontsize = 20)
  plt.xlabel(r'$\Omega$', fontsize = 16)
  plt.ylabel(r'$|X(\Omega)|$', fontsize = 16)
  plt.legend(loc = 'upper right')
  plt.xlim(0, fs/2)
```

```python
plot_dft(x, Fs, samples = 215,l_label = "DFT with 215 samples")
plt.savefig('q3_p1.png')
plt.figure()
plot_dft(x, Fs, samples = 415,l_label = "DFT with 415 samples")
plt.savefig('q3_p2.png')
plt.figure()
plot_dft(x, Fs, samples = 1115,l_label = "DFT with 1115 samples")
plt.savefig('q3_p3.png')
plt.figure()
plot_dft(x, Fs, samples = 1515,l_label = "DFT with 1515 samples")
plt.savefig('q3_p4.png')
plt.figure()
plot_dft(x, Fs, samples = 1915,l_label = "DFT with 1915 samples")
plt.savefig('q3_p5.png')
plt.figure()
```

```python
def plot_dft(n,x,samples=None,l_label = None,clr = 'b',case = None) :
  plt.plot(n,x,clr,label = l_label)
  plt.legend()
  if case == 1 :
    plt.title('Magnitude Spectrum Plot', fontsize = 20)
  else :
    plt.title('Magnitude Spectrum Plot', fontsize = 20)
  plt.xlabel(r'Normalized frequency $f$(cycles/sample)', fontsize = 16)
  plt.ylabel(r'Normalized fft magnitude', fontsize = 16)
  plt.legend(loc = 'upper right')
  plt.xlim(0, 0.5)
  plt.figure()
```

```python
x=np.loadtxt('Exp4Data1.txt', delimiter=',')
original_length=len(x)
print("Original length before padding: {}".format(len(x)))
x=np.hstack((x, np.zeros(100)))
print("New length before padding: {}".format(len(x)))
N = len(x)
X1=fft(sp.windows.hamming(N)*x, N)
plot_dft((1/N)*np.arange(1, N+1), np.abs((X1/original_length)), l_label = r'Magnitude Spectrum Plot of Signal with Hamming window',case = 1)
plt.savefig('q4_p1.png')
X2=fft(sp.windows.boxcar(N)*x, N)
plot_dft((1/N)*np.arange(1, N+1), np.abs((X2/original_length)), l_label = r'Magnitude Spectrum Plot of Signal with Rectangular window', case = 2)
plt.savefig('q4_p2.png')
Peaks1 = sp.find_peaks(np.abs(X1), prominence=0.5)[0]
Peaks2 = sp.find_peaks(np.abs(X2), prominence=15)[0]
print('Frequencies with Hamming window and peak finding: ', Peaks1[0]/N, Peaks1[1]/N)
print('Frequencies with Rectangular window and peak finding: ', Peaks2[0]/N, Peaks2[1]/N)
```
